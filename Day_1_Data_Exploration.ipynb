```python
# Импорт библиотек
import pandas as pd

# Загрузка данных
df = pd.read_csv(r"C:\Users\maksi\OneDrive\Desktop\SMAD\датасеты_исх\Sleep_health_and_lifestyle.csv")  # Проверь имя файла!

# Базовое исследование
print("1. Первые 5 строк:")
display(df.head())
```

```python
print("\n2. Информация о данных:")
display(df.info())
```

```python
print("\n3. Статистика по числовым колонкам:")
display(df.describe())
```

```python
print("\n4. Уникальные значения в колонках:")
print("Профессии:", df['Occupation'].unique())
print("Гендер:", df['Gender'].unique())
```

```python
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
```

```python
display(df.head())
```

```python
df = df.drop('Blood Pressure', axis=1)
```

```python
# BMI Category: Normal/Normal Weight -> 0, Overweight -> 1, Obese -> 2
bmi_mapping = {
    'Normal': 0,
    'Normal Weight': 0,  # Добавляем альтернативное написание
    'Overweight': 1,
    'Obese': 2
}

# Создаем колонку с числовыми кодами
df['BMI_Code'] = df['BMI Category'].map(bmi_mapping)

# Проверяем уникальные значения в исходной колонке
print("Уникальные значения BMI Category:", df['BMI Category'].unique())
```

```python
# Gender: Male -> 0, Female -> 1
df['Gender_Code'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Автоматическое преобразование профессий в числовые коды
df['Occupation_Code'] = df['Occupation'].astype('category').cat.codes

# Создаем словарь для расшифровки
occupation_mapping = dict(enumerate(df['Occupation'].astype('category').cat.categories))
print("Коды профессий:", occupation_mapping)
```

```python
# Создаем бинарный признак (1 - есть нарушение, 0 - нет)
df['Has_Sleep_Disorder'] = df['Sleep Disorder'].apply(
    lambda x: 0 if x == 'None' else 1
)

# Альтернатива для нескольких категорий
disorder_mapping = {
    'None': 0,
    'Insomnia': 1,
    'Sleep Apnea': 2
}
df['Sleep_Disorder_Code'] = df['Sleep Disorder'].map(disorder_mapping)
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
```

```python
import json

mappings = {
    'Gender': {'Male': 0, 'Female': 1},
    'BMI': bmi_mapping,
    'Occupation': occupation_mapping,
    'Sleep Disorder': disorder_mapping
}

with open('column_mappings.json', 'w') as f:
    json.dump(mappings, f)
```

```python
# Выводим типы данных после преобразований
print("\nТипы данных после преобразований:")
print(df.dtypes)

# Пример преобразованных колонок
display(df[['Gender_Code','BMI_Code',
           'Has_Sleep_Disorder']].head())
```

```python
df = df.drop(['Gender', 'BMI Category', 'Occupation', 'Sleep Disorder'], axis=1)
```

```python
# Просто выводим все мэппинги как есть
print("=== ПОЛНЫЙ СПИСОК КОДИРОВОК ===")
print("\nGender_Code (Пол):")
for name, code in mappings['Gender'].items():
    print(f"{code}: {name}")

print("\nBMI_Code (Индекс массы тела):")
for name, code in mappings['BMI'].items():
    print(f"{code}: {name}")

print("\nOccupation_Code (Профессия):")
for code, name in mappings['Occupation'].items():
    print(f"{code}: {name}")

print("\nSleep_Disorder_Code (Нарушения сна):")
for name, code in mappings['Sleep Disorder'].items():
    print(f"{code}: {name}")
```

```python
TARGET_TYPE = 'int64'   # Стандартный вариант (без экономии памяти)

# Список столбцов для преобразования
columns_to_convert = [
    'Person ID', 'Age', 'Sleep Duration', 'Quality of Sleep',
    'Physical Activity Level', 'Stress Level', 'Heart Rate',
    'Daily Steps', 'Systolic_BP', 'Diastolic_BP',
    'Gender_Code', 'BMI_Code', 'Occupation_Code',
    'Has_Sleep_Disorder', 'Sleep_Disorder_Code'
]

# Преобразование типов данных
df[columns_to_convert] = df[columns_to_convert].astype(TARGET_TYPE)

# Проверка результатов
print("Типы данных после преобразования:")
print(df.dtypes)
```

```python
# Убедитесь, что нет потери данных
print("\nМинимальные значения:")
print(df[columns_to_convert].min())

print("\nМаксимальные значения:")
print(df[columns_to_convert].max())
```
