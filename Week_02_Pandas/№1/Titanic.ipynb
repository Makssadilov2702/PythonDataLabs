```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Titanic_Data.csv', sep=';')
df
```

```python
print("\nИнформация о данных:")
display(df.info())
print("\nСтатистика по числовым колонкам:")
display(df.describe())
print("\nУникальные значения в колонках:")
print("Спасение:", df['Survived'].unique())
print("Пол:", df['Sex'].unique())
print("Класс:", df['Passenger Class'].unique())
```

```python
# Добавляем столбец с id (например, от 1 до количества строк)
df.insert(0, 'ID_passenger', range(1, len(df) + 1))
```

```python
# 1. Заменяем запятые на точки
df['Age'] = df['Age'].str.replace(',', '.', regex=False)
df['Passenger Fare'] = df['Passenger Fare'].str.replace(',', '.', regex=False)
# 2. Преобразуем в числа (дробные, float)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Passenger Fare'] = pd.to_numeric(df['Passenger Fare'], errors='coerce')
print("\nТип данных Age:", df['Age'].dtype)
```

```python
def categorize_age(age):
    if pd.isna(age):
        return "Missing"
    elif age <= 1:
        return "Infant(0-1]"
    elif age <= 5:
        return "Toddler (1-5]"
    elif age <= 12:
        return "Child (5-12] " 
    elif age <= 18:
        return "Teen(12-18]"
    elif age <= 30:
        return "Young Adult (18-30]"
    elif age <= 50:
        return "Adult (30-50]"
    elif age <= 65:
        return "Middle Aged (50-65]"
    else:
        return "Senior (65-)"

df['Age_Category'] = df['Age'].apply(categorize_age)
print("\nAge category distribution:")
print(df['Age_Category'].value_counts(dropna=False))
```

```python
def categorize_fare(fare):
    if pd.isna(fare):
        return "Missing"
    elif fare <= 8:
        return "Lowest Fare (0-8]"
    elif fare <= 10.5:
        return "Budget Fare (8-10.5]"
    elif fare <= 22:
        return "Standard Fare (10.5-22]"
    elif fare <= 41.5:
        return "Premium Fare (22-41.5]"
    else:
        return "Luxury Fare (41.5-513]"

# Применяем функцию к столбцу
df['Fare_Category'] = df['Passenger Fare'].apply(categorize_fare)

# Выводим распределение
print("\nPassenger Fare category distribution:")
print(df['Fare_Category'].value_counts(dropna=False).sort_index())
```

```python
full_duplicates = df[df.duplicated(keep=False)]

print(f"Полных дубликатов строк: {len(full_duplicates)}")
if not full_duplicates.empty:
    print("\nПример дубликатов:")
    print(full_duplicates.head())
```

```python
# Показать дубликаты билетов (полные строки)
ticket_duplicates = df[df.duplicated('Ticket Number', keep=False)].sort_values('Ticket Number')

# Посчитать количество повторений каждого билета
ticket_counts = df['Ticket Number'].value_counts()
repeated_tickets = ticket_counts[ticket_counts > 1]

print(f"Уникальных номеров билетов: {df['Ticket Number'].nunique()}")
print(f"Билетов с дубликатами: {len(repeated_tickets)}")
print("\nТоп-5 самых частых билетов:")
print(repeated_tickets.head())
```

```python
survival_column = 'Survived'
df['Survived_num'] = df[survival_column].map({'Yes': 1, 'No': 0})

Общая статистика выживаемости
print("Общая выживаемость:")
print(df[survival_column].value_counts(normalize=True).mul(100).round(1))
```

```python
target_column = 'Survived_num'  # 1-выжил, 0-погиб
categorical_columns = ['Sex', 'Pclass', 'Age_Category', 'Fare_Category'] 

# Открываем файл для записи
with open('analysis_results.txt', 'w', encoding='utf-8') as f:
    # 1. Анализ по категориям
    for col in categorical_columns:
        if col in df.columns:
            stats = df.groupby(col, observed=False)[target_column].agg(
                Выжило='sum',
                Погибло=lambda x: len(x)-x.sum(),
                Всего='count',
                Доля_выживших='mean'
            )
            
            # Вывод на экран
            print(f"\nСтатистика по {col}:")
            print(stats.round(2))
            
            # Запись в файл
            f.write(f"\nСтатистика по {col}:\n")
            f.write(stats.round(2).to_string())  # to_string() для корректного форматирования таблицы
            f.write("\n")  # Добавляем перенос строки

print("Анализ сохранен в файл 'analysis_results.txt'")
```
