# –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö: –∫–æ–Ω—Å–ø–µ–∫—Ç –¥–ª—è –∞–Ω–∞–ª–∏—Ç–∏–∫–∞
**–¶–µ–ª—å**: –≤—ã—è–≤–ª–µ–Ω–∏–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–µ–π, –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑ –∏ –ø—Ä–∏–Ω—è—Ç–∏–µ —Ä–µ—à–µ–Ω–∏–π –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω—ã—Ö.

## –û—Å–Ω–æ–≤–Ω—ã–µ –ø–æ–Ω—è—Ç–∏—è
- **–ì–µ–Ω–µ—Ä–∞–ª—å–Ω–∞—è —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å** ‚Äî –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã/–Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –º—ã —Ö–æ—Ç–∏–º –∏–∑—É—á–∏—Ç—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤—Å–µ –∫–ª–∏–µ–Ω—Ç—ã –±–∞–Ω–∫–∞).
- **–í—ã–±–æ—Ä–∫–∞** ‚Äî –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 1000 —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤).
- **–ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ** ‚Äî —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –∏–∑–º–µ—Ä—è—é—Ç—Å—è:
  - *–ó–∞–≤–∏—Å–∏–º–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è* (Y) ‚Äî —Ç–æ, —á—Ç–æ –º—ã –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º (–Ω–∞–ø—Ä–∏–º–µ—Ä, –æ—Ç—Ç–æ–∫ –∫–ª–∏–µ–Ω—Ç–æ–≤).
  - *–ù–µ–∑–∞–≤–∏—Å–∏–º–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è* (X) ‚Äî —Ñ–∞–∫—Ç–æ—Ä—ã, –≤–ª–∏—è—é—â–∏–µ –Ω–∞ Y (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–æ–∑—Ä–∞—Å—Ç, –¥–æ—Ö–æ–¥).

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏**: 
–ì–µ–Ω–µ—Ä–∞–ª—å–Ω–∞—è —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å ‚Äî —ç—Ç–æ ¬´–≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ¬ª, –≤—ã–±–æ—Ä–∫–∞ ‚Äî ¬´–∫—É—Å–æ—á–µ–∫ –¥–∞–Ω–Ω—ã—Ö, —Å –∫–æ—Ç–æ—Ä—ã–º –º—ã —Ä–µ–∞–ª—å–Ω–æ —Ä–∞–±–æ—Ç–∞–µ–º¬ª. –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ ‚Äî —ç—Ç–æ —Å—Ç–æ–ª–±—Ü—ã –≤ —Ç–∞–±–ª–∏—Ü–µ: –æ–¥–∏–Ω –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç (Y), –¥—Ä—É–≥–∏–µ ‚Äî –ø—Ä–∏—á–∏–Ω—ã (X).

## –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö
- **–ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ** (—á–∏—Å–ª–∞):
  - *–î–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ*: —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫).
  - *–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–µ*: –ª—é–±—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ (–≤–µ—Å, –¥–æ—Ö–æ–¥).
- **–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ** (–≥—Ä—É–ø–ø—ã):
  - *–ù–æ–º–∏–Ω–∞–ª—å–Ω—ã–µ*: –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –±–µ–∑ –ø–æ—Ä—è–¥–∫–∞ (—Ü–≤–µ—Ç–∞, –ø–æ–ª).
  - *–ü–æ—Ä—è–¥–∫–æ–≤—ã–µ*: –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å –ø–æ—Ä—è–¥–∫–æ–º (—É—Ä–æ–≤–µ–Ω—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏: –Ω–∏–∑–∫–∏–π/—Å—Ä–µ–¥–Ω–∏–π/–≤—ã—Å–æ–∫–∏–π).

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏**: 
–ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ ‚Äî —ç—Ç–æ —Ç–æ, —á—Ç–æ –º–æ–∂–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å (¬´5 —è–±–ª–æ–∫¬ª). –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ ‚Äî —ç—Ç–æ –º–µ—Ç–∫–∏ (¬´–∫—Ä–∞—Å–Ω—ã–µ —è–±–ª–æ–∫–∏¬ª, ¬´–∑–µ–ª—ë–Ω—ã–µ —è–±–ª–æ–∫–∏¬ª). –ü–æ—Ä—è–¥–∫–æ–≤—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –º–æ–∂–Ω–æ —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞—Ç—å (¬´–º–∞–ª–µ–Ω—å–∫–∏–π¬ª, ¬´—Å—Ä–µ–¥–Ω–∏–π¬ª, ¬´–±–æ–ª—å—à–æ–π¬ª).

## –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
–ú–µ—Ç–æ–¥—ã –¥–ª—è –∫—Ä–∞—Ç–∫–æ–≥–æ –æ–ø–∏—Å–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö:
- **–ú–µ—Ä—ã —Ü–µ–Ω—Ç—Ä–∞**: 
  - –°—Ä–µ–¥–Ω–µ–µ ($\bar{x} = \frac{\sum x_i}{n}$),
  - –ú–µ–¥–∏–∞–Ω–∞ (—Å–µ—Ä–µ–¥–∏–Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö),
  - –ú–æ–¥–∞ (—Å–∞–º–æ–µ —á–∞—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ).
- **–ú–µ—Ä—ã —Ä–∞–∑–±—Ä–æ—Å–∞**:
  - –î–∏—Å–ø–µ—Ä—Å–∏—è ($\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}$),
  - –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ ($\sigma$),
  - –†–∞–∑–º–∞—Ö (—Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É max –∏ min).

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑
- **H0 (–Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)**: ¬´–Ω–∏—á–µ–≥–æ –Ω–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å¬ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –æ–¥–∏–Ω–∞–∫–æ–≤ –¥–æ –∏ –ø–æ—Å–ª–µ –∞–∫—Ü–∏–∏).
- **H1 (–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)**: ¬´–µ—Å—Ç—å —ç—Ñ—Ñ–µ–∫—Ç¬ª (—Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –≤—ã—Ä–æ—Å).
- **p-value**: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Ç–∞–∫–∏–µ –¥–∞–Ω–Ω—ã–µ, –µ—Å–ª–∏ H0 –≤–µ—Ä–Ω–∞.
- **–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (Œ±)**: –ø–æ—Ä–æ–≥ –¥–ª—è –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è H0 (–æ–±—ã—á–Ω–æ 0.05). –ï—Å–ª–∏ p-value < Œ± ‚Äî –æ—Ç–≤–µ—Ä–≥–∞–µ–º H0.

### üîç –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã
> **–ß—Ç–æ —ç—Ç–æ**: –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π, –≤ –∫–æ—Ç–æ—Ä–æ–º —Å –∑–∞–¥–∞–Ω–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –ª–µ–∂–∏—Ç –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞.
> **–§–æ—Ä–º—É–ª–∞**: $\text{–°—Ä–µ–¥–Ω–µ–µ} \pm z \cdot \frac{\sigma}{\sqrt{n}}$
> **–ü—Ä–∏–º–µ—Ä**: –°—Ä–µ–¥–Ω–∏–π —á–µ–∫ = 5000 —Ä—É–±. (95% –î–ò: [4800, 5200]).

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏**: –≠—Ç–æ ¬´–¥–∏–∞–ø–∞–∑–æ–Ω —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏¬ª. –ï—Å–ª–∏ –ø—Ä–æ–≤–µ—Å—Ç–∏ 100 —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤, –≤ 95 –∏–∑ –Ω–∏—Ö –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ–ø–∞–¥—ë—Ç –≤ —ç—Ç–æ—Ç –∏–Ω—Ç–µ—Ä–≤–∞–ª.

```python
# –ü—Ä–∏–º–µ—Ä —Ä–∞—Å—á—ë—Ç–∞ 95% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –¥–ª—è —Å—Ä–µ–¥–Ω–µ–≥–æ
import scipy.stats as stats
import numpy as np

# –î–∞–Ω–Ω—ã–µ: –≤—ã–±–æ—Ä–∫–∞ –∏–∑ 100 –∫–ª–∏–µ–Ω—Ç–æ–≤
sample = [5000, 5200, 4800, 5100, 4900]  # –ü—Ä–∏–º–µ—Ä –Ω–µ–±–æ–ª—å—à–æ–π –≤—ã–±–æ—Ä–∫–∏
mean = np.mean(sample)
std = np.std(sample, ddof=1)  # ddof=1 –¥–ª—è –Ω–µ—Å–º–µ—â—ë–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏
n = len(sample)

# –†–∞—Å—á—ë—Ç –î–ò
z = stats.norm.ppf(0.975)  # z-–∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è 95% –î–ò
ci_lower = mean - z * (std / np.sqrt(n))
ci_upper = mean + z * (std / np.sqrt(n))

print(f"95% –î–ò: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

### ‚ö†Ô∏è –û—à–∏–±–∫–∏ I –∏ II —Ä–æ–¥–∞
| –û—à–∏–±–∫–∞ | –û–ø–∏—Å–∞–Ω–∏–µ | –ö–∞–∫ –∏–∑–±–µ–∂–∞—Ç—å |
|--------|----------|--------------|
| **I —Ä–æ–¥–∞ (Œ±)** | –õ–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç (–æ—Ç–≤–µ—Ä–≥–ª–∏ H0, –∫–æ–≥–¥–∞ –æ–Ω–∞ –≤–µ—Ä–Ω–∞) | –£–º–µ–Ω—å—à–∏—Ç—å Œ± (–Ω–æ —ç—Ç–æ —É–≤–µ–ª–∏—á–∏—Ç —Ä–∏—Å–∫ –æ—à–∏–±–∫–∏ II —Ä–æ–¥–∞) |
| **II —Ä–æ–¥–∞ (Œ≤)** | –õ–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç (–Ω–µ –æ—Ç–≤–µ—Ä–≥–ª–∏ H0, –∫–æ–≥–¥–∞ –æ–Ω–∞ –ª–æ–∂–Ω–∞) | –£–≤–µ–ª–∏—á–∏—Ç—å —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –∏–ª–∏ –º–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞ |

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏**:  
- –û—à–∏–±–∫–∞ I —Ä–æ–¥–∞: ¬´–ë–æ–ª—å–Ω–æ–π –∑–¥–æ—Ä–æ–≤¬ª (—Ç–µ—Å—Ç –ø–æ–∫–∞–∑–∞–ª, —á—Ç–æ –ª–µ–∫–∞—Ä—Å—Ç–≤–æ —Ä–∞–±–æ—Ç–∞–µ—Ç, –Ω–æ —ç—Ç–æ –Ω–µ —Ç–∞–∫).  
- –û—à–∏–±–∫–∞ II —Ä–æ–¥–∞: ¬´–ó–¥–æ—Ä–æ–≤ –±–æ–ª—å–Ω–æ–π¬ª (—Ç–µ—Å—Ç –Ω–µ –∑–∞–º–µ—Ç–∏–ª —ç—Ñ—Ñ–µ–∫—Ç, —Ö–æ—Ç—è –æ–Ω –µ—Å—Ç—å).

## –û—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã
- **t-—Ç–µ—Å—Ç**: —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö 2 –≥—Ä—É–ø–ø (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–æ—Ö–æ–¥ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω).
- **ANOVA**: —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö ‚â•3 –≥—Ä—É–ø–ø (–Ω–∞–ø—Ä–∏–º–µ—Ä, —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å 3 –ª–µ–∫–∞—Ä—Å—Ç–≤).
- **œá¬≤-—Ç–µ—Å—Ç**: –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å–≤—è–∑–∏ –º–µ–∂–¥—É –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø–æ–ª–∞ –æ—Ç –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è —Ç–æ–≤–∞—Ä–∞).

### üìå –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–π —Ç–µ—Å—Ç–æ–≤
- **–ù–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å**: —Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –∏–ª–∏ QQ-plot.
- **–ì–æ–º–æ—Å–∫–µ–¥–∞—Å—Ç–∏—á–Ω–æ—Å—Ç—å**: —Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞.
- **–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä —è—á–µ–π–∫–∏ (œá¬≤)**: –æ–∂–∏–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã > 5.

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ (Shapiro-Wilk)
from scipy import stats

data = [5000, 5200, 4800, 5100, 4900]
stat, p = stats.shapiro(data)
print(f"p-value –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏: {p:.4f}")
if p > 0.05:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ (H0 –Ω–µ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞)")
else:
    print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ù–ï –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–æ–º–æ—Å–∫–µ–¥–∞—Å—Ç–∏—á–Ω–æ—Å—Ç–∏ (Levene)
group1 = [5000, 5200, 4800]
group2 = [4900, 5100, 5050]
stat, p = stats.levene(group1, group2)
print(f"\np-value –≥–æ–º–æ—Å–∫–µ–¥–∞—Å—Ç–∏—á–Ω–æ—Å—Ç–∏: {p:.4f}")
```

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏**:  
t-—Ç–µ—Å—Ç ‚Äî ¬´–∫—Ç–æ –∫—Ä—É—á–µ –∏–∑ –¥–≤–æ–∏—Ö¬ª, ANOVA ‚Äî ¬´–∫—Ç–æ –∫—Ä—É—á–µ –∏–∑ —Ç—Ä–æ–∏—Ö¬ª, œá¬≤-—Ç–µ—Å—Ç ‚Äî ¬´–∑–∞–≤–∏—Å–∏—Ç –ª–∏ –≤—ã–±–æ—Ä —Ü–≤–µ—Ç–∞ –æ—Ç –ø–æ–ª–∞?¬ª.

## –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∏ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è
- **–ö–æ–≤–∞—Ä–∏–∞—Ü–∏—è**: –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Å–≤—è–∑–∏ –º–µ–∂–¥—É –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ (–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è/–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–∞—è).
- **–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è (–ü–∏—Ä—Å–æ–Ω)**: —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è (-1 ‚â§ r ‚â§ 1):
  - $r = 0$: —Å–≤—è–∑–∏ –Ω–µ—Ç,
  - $r > 0$: –ø—Ä—è–º–∞—è —Å–≤—è–∑—å,
  - $r < 0$: –æ–±—Ä–∞—Ç–Ω–∞—è —Å–≤—è–∑—å.
- **–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –°–ø–∏—Ä–º–µ–Ω–∞**: –¥–ª—è –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö —Å–≤—è–∑–µ–π (—Ä–∞–±–æ—Ç–∞–µ—Ç —Å —Ä–∞–Ω–≥–∞–º–∏).

```python
# –†–∞—Å—á—ë—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ –∏ –°–ø–∏—Ä–º–µ–Ω–∞
import pandas as pd

df = pd.DataFrame({
    '–í–æ–∑—Ä–∞—Å—Ç': [25, 30, 35, 40, 45],
    '–¢—Ä–∞—Ç—ã': [10000, 12000, 15000, 13000, 11000]
})

pearson = df['–í–æ–∑—Ä–∞—Å—Ç'].corr(df['–¢—Ä–∞—Ç—ã'], method='pearson')
spearman = df['–í–æ–∑—Ä–∞—Å—Ç'].corr(df['–¢—Ä–∞—Ç—ã'], method='spearman')

print(f"–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ü–∏—Ä—Å–æ–Ω–∞: {pearson:.2f}")
print(f"–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –°–ø–∏—Ä–º–µ–Ω–∞: {spearman:.2f}")
```

## –ö–∞–∫ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–µ—Å—Ç–æ–≤
1. **p-value < Œ± (0.05)**: –æ—Ç–≤–µ—Ä–≥–∞–µ–º H0, —ç—Ñ—Ñ–µ–∫—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.
2. **p-value ‚â• Œ±**: –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º H0, –¥–∞–Ω–Ω—ã—Ö –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª—è –≤—ã–≤–æ–¥–æ–≤.
3. **–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è**:
   - |r| > 0.7: —Å–∏–ª—å–Ω–∞—è —Å–≤—è–∑—å,
   - 0.3 < |r| < 0.7: —É–º–µ—Ä–µ–Ω–Ω–∞—è,
   - |r| < 0.3: —Å–ª–∞–±–∞—è.

**–í–∞–∂–Ω–æ**: –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è ‚â† –ø—Ä–∏—á–∏–Ω–Ω–æ-—Å–ª–µ–¥—Å—Ç–≤–µ–Ω–Ω–∞—è —Å–≤—è–∑—å!

### üìè –†–∞–∑–º–µ—Ä —ç—Ñ—Ñ–µ–∫—Ç–∞
> **–ó–∞—á–µ–º**: p-value –≥–æ–≤–æ—Ä–∏—Ç –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –Ω–æ –Ω–µ –æ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–π –≤–∞–∂–Ω–æ—Å—Ç–∏.
> - **–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ö–æ–±–µ–Ω–∞ (d)**: $d = 0.2$ ‚Äî —Å–ª–∞–±—ã–π, $d = 0.5$ ‚Äî —É–º–µ—Ä–µ–Ω–Ω—ã–π, $d = 0.8$ ‚Äî —Å–∏–ª—å–Ω—ã–π.
> - **–≠—Ç–∞-–∫–≤–∞–¥—Ä–∞—Ç (Œ∑¬≤)**: –¥–æ–ª—è –æ–±—ä—è—Å–Ω—ë–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏.

```python
# –†–∞—Å—á—ë—Ç —Ä–∞–∑–º–µ—Ä–∞ —ç—Ñ—Ñ–µ–∫—Ç–∞ (–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ö–æ–±–µ–Ω–∞)
def cohen_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    s1, s2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

group1 = [5000, 5200, 4800]
group2 = [4900, 5100, 5050]
d = cohen_d(group1, group2)
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ö–æ–±–µ–Ω–∞: {d:.2f}")
```

### üîÅ –ú–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑
> **–ü—Ä–æ–±–ª–µ–º–∞**: –ü—Ä–∏ 20 —Ç–µ—Å—Ç–∞—Ö —Å Œ±=0.05 –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ = 64%.
> **–†–µ—à–µ–Ω–∏–µ**:
> - –ö–æ—Ä—Ä–µ–∫—Ü–∏—è –ë–æ–Ω—Ñ–µ—Ä—Ä–æ–Ω–∏: $\alpha_{\text{–Ω–æ–≤—ã–π}} = \frac{\alpha}{k}$
> - –ú–µ—Ç–æ–¥ –•–æ–ª–º–∞: –ø–æ—à–∞–≥–æ–≤–∞—è –∫–æ—Ä—Ä–µ–∫—Ü–∏—è.

```python
# –ö–æ—Ä—Ä–µ–∫—Ü–∏—è –ë–æ–Ω—Ñ–µ—Ä—Ä–æ–Ω–∏
p_values = [0.01, 0.04, 0.06, 0.001]
alpha = 0.05
bonferroni_alpha = alpha / len(p_values)

significant = [p < bonferroni_alpha for p in p_values]
print(f"–ö–æ—Ä—Ä–µ–∫—Ü–∏—è –ë–æ–Ω—Ñ–µ—Ä—Ä–æ–Ω–∏ (Œ±={bonferroni_alpha:.4f})")
for i, p in enumerate(p_values):
    print(f"–¢–µ—Å—Ç {i+1}: p={p:.3f}, –∑–Ω–∞—á–∏–º? {significant[i]}")
```
